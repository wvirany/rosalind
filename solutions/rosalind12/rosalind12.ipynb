# Overlap Graphs

---

## Problem:

`Given`: A collection of DNA strings in FASTA format having total length at most 10 kbp.

`Return`: The adjacency list corresponding to $O_3$.

```python
# Parsing Data

data = open('rosalind_grph.txt', 'r')
raw = data.read()

id = []
d = []

for seqblock in raw.split(">")[1:]:

    parts = seqblock.split("\n")

    fasta = parts[0]
    seq = ''.join(parts[1:])

    id.append(fasta)
    d.append(seq)

data.close()
```

```python
adj_list = []

for i, j in zip(id, d):
  for k, l in zip(id, d):
    if j[-3:] == l[:3] and i != k: adj_list.append((i,k))
```

```python
for i in adj_list:
  print(i[0], i[1])
```
